```python
df = df.drop_duplicates()
```

```python
num_cols = df.select_dtypes(include="number").columns
df[num_cols] = df[num_cols].fillna(df[num_cols].median())
```

```python
cat_cols = df.select_dtypes(include="object").columns
df[cat_cols] = df[cat_cols].fillna("Unknown")
```

```python
df["check_in_time"] = pd.to_datetime(df["check_in_time"])
df["consultation_start_time"] = pd.to_datetime(df["consultation_start_time"])
df["check_out_time"] = pd.to_datetime(df["check_out_time"])
df["visit_date"] = pd.to_datetime(df["visit_date"])
```

```python
df["waiting_time_minutes"] = (
    df["consultation_start_time"] - df["check_in_time"]
).dt.total_seconds() / 60
df["consultation_duration_minutes"] = (
    df["check_out_time"] - df["consultation_start_time"]
).dt.total_seconds() / 60
df["doctor_utilization_rate"] = (
    df["doctor_worked_minutes"] / df["doctor_available_minutes"]
)
```

```python
df = df[df["waiting_time_minutes"] <= 240]
df = df[df["doctor_utilization_rate"] <= 1]
```

```python
df.columns = df.columns.str.lower()
```

```python
df.to_csv("Hospital_visits_clean.csv", index=False)
```

```python
from sqlalchemy import create_engine
import pandas as pd

engine = create_engine(
    "postgresql+psycopg2://postgres:123456@localhost:5432/hospital_management"
)
```

```python
df = pd.read_csv("Hospital_visits_clean.csv")
```

```python
dim_date = df[['visit_date']].drop_duplicates()
dim_date['visit_date'] = pd.to_datetime(dim_date['visit_date'])

dim_date['year'] = dim_date['visit_date'].dt.year
dim_date['month'] = dim_date['visit_date'].dt.month
dim_date['month_name'] = dim_date['visit_date'].dt.month_name()
dim_date['day'] = dim_date['visit_date'].dt.day
dim_date['weekday'] = dim_date['visit_date'].dt.day_name()

dim_date.rename(columns={'visit_date': 'date_id'}, inplace=True)

dim_date.to_sql('dim_date', engine, if_exists='append', index=False)
```

```python
dim_department = (
    df[['department', 'available_beds']]
    .groupby('department', as_index=False)
    .mean()
    .rename(columns={'department': 'department_name'})
)

dim_department.to_sql('dim_department', engine, if_exists='append', index=False)
```

```python
dim_shift = (
    df[['shift']]
    .drop_duplicates()
    .rename(columns={'shift': 'shift_name'})
)

dim_shift.to_sql('dim_shift', engine, if_exists='append', index=False)
```

```python
dim_doctor = (
    df[['doctor_id', 'department']]
    .groupby('doctor_id', as_index=False)
    .first()
    .rename(columns={'department': 'department_name'})
)

dim_doctor.to_sql('dim_doctor', engine, if_exists='append', index=False)
```

```python
dept = pd.read_sql("SELECT * FROM dim_department", engine)
shift = pd.read_sql("SELECT * FROM dim_shift", engine)

df_fact = df.merge(dept, left_on='department', right_on='department_name')
df_fact = df_fact.merge(shift, left_on='shift', right_on='shift_name')
```

```python
fact = df_fact[[
    'appointment_id',
    'patient_id',
    'doctor_id',
    'department_id',
    'shift_id',
    'visit_date',
    'waiting_time_minutes',
    'consultation_duration_minutes',
    'patient_volume_shift',
    'patients_seen_by_doctor_shift',
    'doctor_available_minutes',
    'doctor_worked_minutes',
    'doctor_utilization_rate',
    'satisfaction_score'
]]

fact.rename(columns={'visit_date': 'date_id'}, inplace=True)

fact.to_sql('fact_patient_visit', engine, if_exists='append', index=False)
```
